# 2  计算每个股票每分钟已实现的日内回报

以每只股票9:31:00的close(=mark_price)价作为基准价，计算股票的每分钟的回报率

每分钟的回报为：

$$
 return\_of\_stock= \frac{close-mark\_price}{mark\_price}
$$

```python

stock_return_data=data.iloc[:,:-2]
stock_return_data[:2]
```

```python
def individual_stock(stock_code):
    stock_code=int(stock_code)
    return stock_return_data[stock_return_data.stk_d_code==stock_code]

stock_code=stock_return_data.stk_d_code.values
stock_code=list(set(stock_code))
```

```python
def stock_return(stock_code):
    individual_stock_data=individual_stock(stock_code)
    mark_price=individual_stock_data.iloc[0,-1]
    individual_stock_data['return_rate']=individual_stock_data.close.apply(lambda x : (x-mark_price)/mark_price)
    return individual_stock_data
```

## 所有股票的回报为

```python
all_return=map(stock_return,stock_code)
all_return=[i for i in all_return]
```

# 3 计算行业均价指数(行业中的股票价格算术平均)，从1000起始

以close的均值为mark_mean价格为准,9:31:00的均值为 1000

$$
index\_of\_industry= \frac{mean\_of\_all\_stock\_closes}{mark\_mean}*1000
$$

```python
industry_data=orignal_data.copy()

# industry_data=industry_data[:2400]

industry_data[:2]
```

```python
industry_all_code=list(set(industry.stk_d_swcode))
len(industry_all_code)
```

```python
def individual_industry(industry_code):
    return data_industry[data_industry.stk_d_swcode.isin([industry_code])]


def industry_index(industry_code):
    temp=individual_industry(industry_code).groupby(by='time',sort=False).mean()
    makr_mean=temp.close[0]
    temp['industry_index']=temp.close.apply(lambda x :(x/makr_mean)*10000 )
    return temp
```

## 所有28个行业的指数

# 4 计算每个行业每分钟的金融属性，并且编码：
* mean_20 = 20分钟均线位置(当前的行业指数点位/20分钟的均线点位-1）
* range_20 = 20分钟区间位置(当天的行业指数-20分钟内行业指数最小)/(20分钟内行业指数最大-20分钟内行业指数最小)
* 20分钟beta(采用经典算法）

```python
def get_one_industry(industry_code):
    return all_industry_index[industry_all_code.index(industry_code)][['close','industry_index']]
```

```python
def industry_property(industry_code):
    one_industry=get_one_industry(industry_code)
    temp_20m=get_one_industry(industry_code).rolling(20).mean().industry_index
    one_industry['temp_20m']=temp_20m
    one_industry['mean_20']=(one_industry.industry_index/one_industry.temp_20m)-1
    max_20=one_industry.industry_index.rolling(20).max()
    min_20=one_industry.industry_index.rolling(20).min()

    one_industry['max_20']=max_20
    one_industry['min_20']=min_20

    range_20=(one_industry.industry_index-min_20)/(max_20-min_20)
    one_industry['range_20']=range_20
    return one_industry[['industry_index','mean_20','range_20']]

```

## 所有28个行业的行业属性

```python
all_industry_property=map(industry_property,industry_all_code)
all_industry_property= [i for i in all_industry_property]
```

## 20 munites beta of the industry  （20分钟的beta）

```python
print(industry_all_code)
```

```python
def get_one_industry_simple(industry_code):
    temp=get_one_industry(industry_code).industry_index
    temp=temp.rename(industry_code)
    return temp

all_temp=[]
for i in industry_all_code:
    all_temp.append(get_one_industry_simple(i))
all_industry=pd.DataFrame(all_temp).T
```

## 所有行业的指数每分钟的指数前5行

```python
all_industry[:5]  
```

$$
{\displaystyle \beta ={\frac {\mathrm {Cov} (r_{a},r_{b})}{\mathrm {Var} (r_{b})}},}
r_a为指数,r_b为20分钟的指数均值
$$



```python
def beta_industry(industry_name):
    temp=get_one_industry(industry_name)['industry_index']
    temp=pd.DataFrame(temp.rename(industry_name))
    temp['20_mean_index']=temp.rolling(20).mean()
    temp=temp.dropna()
    cov=temp.cov().iloc[0,1]
    var=temp.cov().iloc[1,1]
    bate=cov/var
    return bate

all_beta=map(beta_industry,industry_all_code)
all_beta=[i for i in all_beta]
```

```python
def dec_industry(time):
    x=all_industry[all_industry.index.isin([time])].T
    x.index
    values=[float(i) for i in x.values]
    temp=pd.DataFrame({'industry':x.index,'val':values})
    return temp.sort_values(by='val',ascending=False)
```

```python
time_10=dec_industry(time[0])  # 10 点

time_11=dec_industry(time[1])   # 11点 

time_14=dec_industry(time[2])  # 14点

```
